# Traning and Visualizing a Decision Tree

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
iris.data
````

````python
X = iris.data[:, 2:]
X
````

````python
y = iris.target
y
````

````python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
````

````python
from sklearn.tree import export_graphviz
export_graphviz(
    tree_clf,
    out_file=('./iris_tree.dot'),
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True
)
````

Use dot command to generate the image.

```bash
dot -Tpng iris_tree.dot -o iris_tree.png
```

# Maching Predictions

The Decision Tree should output the following probabilities: 0% for Iris setosa (0/54), 90.7% for Iris versicolor (49/54), and 9.3% for Iris virginica (5/54). 

````python
tree_clf.predict_proba([[5, 1.5]])
````

# Instability

Decision Trees love orthogonal decision boundaries, which makes them sensitive to training set rotation.

Decision Trees are very sensitive to small variations in the training data.
